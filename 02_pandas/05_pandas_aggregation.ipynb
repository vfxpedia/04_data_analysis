### 집계 함수


```python
# DataFrame 레벨
print(df.count(), n) # 데이터 개수
print(df.sum(), n) # 합계
# print(df.mean(), n) # 평균 ERROR not numeric
# print(df.median(), n) # 중위값 ERROR not numeric
print(df.mode(), n) # 최빈값
# print(df.argmax(), n) # 최빈값의 인덱스 
print(df.max(), n) # 최대값
print(df.min(), n) # 최소값
```

```python
# Series 레벨

print(df['salary'].count(), n) # 데이터 개수
print(df['salary'].sum(), n) # 합계
print(df['salary'].mean(), n) # 평균
print(df['salary'].median(), n) # 중위값
print(df['salary'].mode(), n) # 최빈값
print(df['salary'].argmax(), n) # 최빈값의 인덱스
print(df['salary'].max(), n) # 최대값
print(df['salary'].min(), n) # 최소값
print(df['salary'].std(), n) # 표준편차
print(df['salary'].var(), n) # 분산
```

- DataFrameGroupBy 객체 반환

```python
df.groupby('department')['salary'].sum()
```

```python
df.groupby('department')
```

```python
groupby_dept = df.groupby('department')
print(type(groupby_dept), n)

print(groupby_dept.size(), n)
print(groupby_dept.sum(), n)
print(groupby_dept.groups, n)
print(groupby_dept.get_group('IT'), n)




```

```python
# 부서로 그룹핑 되어있는 데이터 프레임이 가지고 있는 샐러리의 평균값을 구해주세요.
# 그룹별 집계 함수 조회 - 숫자형 데이터에서만 사용가능

print(df.groupby('department')['salary'].mean(), n)
print(df.groupby('department')['salary'].max(), n)
print(df.groupby('department')['salary'].min(), n)
print(df.groupby('department')['salary'].min().values, n)
```

```python
# named aggregation -> 집계 함수 이름 지정 가능, 여러개의 집계 함수 적용 가능
# 딕셔너리 형태로 적용, 키는 컬럼명, 값은 집계 함수
# 하나의 표로 묶어서 보여줄 수 있음, -> 활용 가능한 범위가 많겠는데?

groupby_dept.agg(
        salary_mean = ('salary', 'mean'),
        salary_max = ('salary', 'max'),
        hire_date_min = ('hire_date', 'min'),
        hire_date_max = ('hire_date', 'max')
)
```

#### pivot_table
- 특정 컬럼(인덱스, 컬럼)을 기준으로 값 컬럼을 열람하는 기능
- 연관 있는 데이터를 가독성 좋게 출력 가능

```python
data = {
    'Date': ['2024-10-11', '2024-10-12', '2024-10-13', '2024-10-14', '2024-10-14'],
    'City': ['서울', '인천', '평택', '화성', '서울'],
    'Temperature': [17, 18, 19, 18, 19]
}
df = pd.DataFrame(data)
df
```

```python
# df.T 전치 행렬
# df.pivot_table 피벗 테이블
# index : 행 인덱스로 사용할 컬럼
# columns : 열 인덱스로 사용할 컬럼
# values : 데이터로 사용할 컬럼
# aggfunc : 집계 함수
# fill_value : 결측치 처리 값
# margins : 총계 여부

# ! 왜 피벗 테이블으로 변환하는 것이 중요한지 알아보자.
# 피벗 테이블은 데이터를 효과적으로 분석하고 시각화하기 위해 사용되는 표 형식의 데이터 구조
# 피벗 테이블을 사용하면 데이터를 쉽게 회전시키고, 집계 함수를 적용하고, 데이터를 분석할 수 있음
```

```python
df.pivot_table(index='Date', columns='City', values='Temperature')
```

```python
data = {
    '날짜': [
        '2024-01-01', '2024-01-01', '2024-01-01', '2024-01-01',
        '2024-01-02', '2024-01-02', '2024-01-02', '2024-01-02',
        '2024-01-03', '2024-01-03'
    ],
    '상품': [
        '아이폰 14', '갤럭시 S23', '아이패드 프로', '아이폰 14',
        '아이폰 14', '갤럭시 S23', '아이패드 프로', '갤럭시 S23',
        '맥북 프로', '아이폰 14'
    ],
    '판매량': [15, 30, 5, 10, 20, 25, 10, 5, 8, 12],
    '가격': [1000, 900, 800, 1000, 1000, 900, 800, 900, 2000, 1000]
}
df = pd.DataFrame(data)
df
```

```python
df.pivot_table(
    index='날짜',
    columns='상품',
    values='판매량',
    fill_value=0, # 결측치 처리 값
    aggfunc='mean' # 중복되는 값에 대한 집계 함수
)
```

```python
df.pivot_table(
    index='날짜',
    columns='상품',
    values='판매량',
    fill_value=0, # 결측치 대체 처리 값
    aggfunc='sum' # 중복되는 값에 대한 집계 함수
)
```

```python
df.pivot_table(
    index='날짜',
    columns='상품',
    values='판매량',
    fill_value=0, # 결측치 대체 처리 값
    aggfunc='sum', # 중복되는 값에 대한 집계 함수
    margins=True # 총계 여부 (행 방향과 열 방향의 합계를 추가)
)
```
