```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 28],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'San Francisco'],
    'Salary': [70000, 80000, 120000, 90000, 75000]
}

df = pd.DataFrame(data)
df
```

```python
# 2. **열과 행 선택**
#     - `Name`과 `City` 열을 선택하여 출력하세요.
df[['Name', 'City']]
```

```python
df.loc[:,['Name','City']]
```

```python
#     - 나이가 30 이상인 행을 선택하여 출력하세요.
df[df['Age'] >= 30]
```

```python
#     - 나이가 30 이상인 행을 선택하여 출력하세요.
df.loc[df['Age'] >= 30]
```

```python
# 3. **행과 열 추가**
#     - `Country` 열을 추가하고, 모두 'USA'로 설정하세요.
#     - 새로운 행을 추가하세요. (Name: Frank, Age: 33, City: Seattle, Salary: 85000)
Country = ['USA'] * len(df)
df['Country'] = Country
df
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 28],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'San Francisco'],
    'Salary': [70000, 80000, 120000, 90000, 75000]
}

df = pd.DataFrame(data)
df
```

```python
# 3. **행과 열 추가**
#     - `Country` 열을 추가하고, 모두 'USA'로 설정하세요.
Country = ['USA'] * len(df)
add_Country_df = pd.DataFrame(Country, columns=['Country'])
df = pd.concat([df, add_Country_df], axis=1)
df
```

```python
# #     - 새로운 행을 추가하세요. (Name: Frank, Age: 33, City: Seattle, Salary: 85000)
add_row_df = pd.DataFrame([{'Name': 'Frank',
                          'Age': 33,    
                          'City': 'Seattle',
                          'Salary': 85000,
                          'Country': 'USA'
                          }])
add_row_df
df = pd.concat([df, add_row_df], axis=0, ignore_index=True)
df
```

```python
# 4. **데이터 삭제**
#     - `Salary` 열을 삭제하세요.
#     - 'David'가 있는 행을 삭제하세요.
df = df.drop('Salary', axis=1)
df
```

```python
df = df[df['Name'] != 'David']
df
```

```python
df = df.drop(df[df['Name'] == 'David'].index)
df
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 28],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'San Francisco'],
    'Salary': [70000, 80000, 120000, 90000, 75000]
}

df = pd.DataFrame(data)
df
```

```python
# 5. **데이터 수정**
#     - `Age`가 35 이상인 사람들의 `Salary`를 5% 증가시키세요.

df.loc[df['Age'] >= 35, 'Salary'] = df.loc[df['Age'] >= 35, 'Salary'] * 1.05
df
```

```python
#     # - 'Alice'의 `City`를 'Boston'으로 수정하세요.
df.loc[df['Name'] == 'Alice', 'City'] = 'Boston'
df
```

```python
df['Age']
```

```python
# 6. **데이터 타입 변환**
#     - `Age`와 `Salary`의 데이터 타입을 float로 변환하세요.
df['Age'] = df['Age'].astype(float)
df['Salary'] = df['Salary'].astype(float)
df
```
