```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 28],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'San Francisco'],
    'Salary': [70000, 80000, 120000, 90000, 75000]
}

df = pd.DataFrame(data)
df
```

```python
#     - 나이가 30 이상인 행을 선택하여 출력하세요.
df.loc[df['Age'] >= 30]
```

```python
# 3. **행과 열 추가**
#     - `Country` 열을 추가하고, 모두 'USA'로 설정하세요.
#     - 새로운 행을 추가하세요. (Name: Frank, Age: 33, City: Seattle, Salary: 85000)
Country = ['USA'] * len(df)
df['Country'] = Country
df
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 28],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'San Francisco'],
    'Salary': [70000, 80000, 120000, 90000, 75000]
}

df = pd.DataFrame(data)
df
```

```python
# 3. **행과 열 추가**
#     - `Country` 열을 추가하고, 모두 'USA'로 설정하세요.
Country = ['USA'] * len(df)
add_Country_df = pd.DataFrame(Country, columns=['Country'])
df = pd.concat([df, add_Country_df], axis=1)
df
```

```python
# #     - 새로운 행을 추가하세요. (Name: Frank, Age: 33, City: Seattle, Salary: 85000)
add_row_df = pd.DataFrame([{'Name': 'Frank',
                          'Age': 33,    
                          'City': 'Seattle',
                          'Salary': 85000,
                          'Country': 'USA'
                          }])
add_row_df
df = pd.concat([df, add_row_df], axis=0, ignore_index=True)
df
```

```python
# 4. **데이터 삭제**
#     - `Salary` 열을 삭제하세요.
#     - 'David'가 있는 행을 삭제하세요.
df = df.drop('Salary', axis=1)
df
```

```python
df = df[df['Name'] != 'David']
df
```

```python
df = df.drop(df[df['Name'] == 'David'].index)
df
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 28],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'San Francisco'],
    'Salary': [70000, 80000, 120000, 90000, 75000]
}

df = pd.DataFrame(data)
df
```

```python
# 5. **데이터 수정**
#     - `Age`가 35 이상인 사람들의 `Salary`를 5% 증가시키세요.

df.loc[df['Age'] >= 35, 'Salary'] = df.loc[df['Age'] >= 35, 'Salary'] * 1.05
df
```

```python
#     # - 'Alice'의 `City`를 'Boston'으로 수정하세요.
df.loc[df['Name'] == 'Alice', 'City'] = 'Boston'
df
```

```python
df['Age']
```

```python
# 6. **데이터 타입 변환**
#     - `Age`와 `Salary`의 데이터 타입을 float로 변환하세요.
df['Age'] = df['Age'].astype(float)
df['Salary'] = df['Salary'].astype(float)
df
```

```python
df.filter(like='S')
# filter(like='S')는 컬럼 이름에 'S'가 포함된 컬럼만 선택해서 DataFrame을 반환합니다.
# 만약 컬럼 이름에 'S'가 포함된 것이 없다면, 빈 DataFrame이 반환되고,
# 이때 원래 DataFrame의 인덱스만 남게 됩니다.
# 따라서 결과에 index 라인만 보이는 것입니다.
# filter(like='')는 DataFrame의 컬럼명(열 이름)에 해당 문자열이 포함되어 있는지 확인해서, 
# 그 컬럼들만 선택하는 기능입니다. 즉, 값이 아니라 "컬럼명"에만 적용됩니다.
# 예시: df.filter(like='S')는 컬럼명에 'S'가 들어간 컬럼만 반환합니다.

# df.values는 numpy array이기 때문에 filter 메서드를 사용할 수 없습니다.
# 만약 각 값에 대해 특정 문자가 포함되어 있는지 찾고 싶다면, 
# DataFrame 전체에 대해 문자열 메서드를 적용하거나, 
# 특정 컬럼에 대해 .str.contains() 등을 사용해야 합니다.

# 예시:
# df[df['Last Name'].str.contains('S')]  # 'Last Name' 값에 'S'가 포함된 행 선택

# 결론: filter(like='')는 컬럼명에만 적용되고, 값에는 직접 사용할 수 없습니다.
```
