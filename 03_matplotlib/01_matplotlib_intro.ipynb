### matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### 기본 사용법

```python
x = [1, 2, 3]
y = [4, 5, 6]
plt.plot(x, y)
plt.show()
```

##### Figure | Axes

 - Figure : 전체 그림이 그려지는 컨테이너너
 - Axes : 그래프가 실제로 그려지는 영역(좌표, 데이터)

```python
fig = plt.figure(figsize=(10, 4), facecolor='green')

ax = plt.gca() # 그래프가 그려지는 영역
ax.plot([1, 2, 10], [4, 5, 6])
print(fig)
print(ax)
plt.title('Hello Matplotlib')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=7, figsize=(10, 6))
# 그래프 그려질 영역에 대한 행과 열의 개수를 지정

print(fig)
print(ax)
ax[0, 0].plot([1, 2, 3], [4, 5, 6])
ax[0, 1].plot([1, 2, 3], [4, 4, 5])
ax[0, 2].plot([1, 2, 3], [4, 3, 2])
ax[0, 3].plot([1, 2, 3], [4, 2, 1])
ax[0, 4].plot([1, 2, 3], [4, 1, 2])
ax[0, 5].plot([1, 2, 3], [4, 0, 3])
ax[0, 6].plot([1, 2, 3], [4, 0, 3])
ax[1, 0].plot([1, 2, 3], [4, 0, 3])
ax[1, 1].plot([1, 2, 3], [4, 0, 3])
ax[1, 2].plot([1, 2, 3], [4, 0, 3])
ax[1, 3].plot([1, 2, 3], [4, 0, 3])
ax[1, 4].plot([1, 2, 3], [4, 0, 3])
ax[1, 5].plot([1, 2, 3], [4, 0, 3]) 
ax[1, 6].plot([1, 2, 3], [4, 0, 3])
plt.show()
```
