##### plot 메서드의 매개변수

```python
plt.figure(figsize=(6, 4))

plt.plot(
    x,
    y,
    color='red', # 선의 색상, red, blue, green, yellow, cyan, magenta, black, white
    marker='o', # 점의 모양, o, +, *, s, d, ^, v, >, <, p, h, 1, 2, 3, 4, x, y, t, h, 
    markersize=10, # 점의 크기, 숫자가 클수록 크게 표시
    markerfacecolor='blue', # 점의 색상, 점의 모양이 있을 때 점의 색상
    markeredgecolor='green', # 점의 테두리 색상, 점의 모양이 있을 때 점의 테두리 색상
    markeredgewidth=2, # 점의 테두리 두께, 숫자가 클수록 두께가 높아짐
    linestyle='--', # 선의 스타일, solid, dashed, dotted, dashdot
    linewidth=2, # 선의 두께, 숫자가 클수록 두께가 높아짐
    alpha=0.5, # 선의 투명도, 0~1 사이의 값, 0이면 완전 투명, 1이면 완전 불투명
    label='y = x^2', # 선의 레이블, 범례에 표시될 이름
    # zorder=10, # 선의 순서, 숫자가 클수록 위에 표시
    # visible=True, # 선의 표시 여부, True이면 표시, False이면 표시하지 않음
    animated=False, # 선의 애니메이션 여부, True이면 애니메이션 표시, False이면 애니메이션 표시하지 않음
    # picker=None, # 선의 선택 여부, None이면 선택 안됨, 숫자이면 선택 가능
    # rasterized=False, # 선의 레스터화 여부, True이면 레스터화, False이면 레스터화 하지 않음
    # snap=None, # 선의 스냅 여부, None이면 스냅 안됨, 숫자이면 스냅 됨
    # solid_capstyle='butt', # 선의 끝 모양, butt, round, projecting
    # solid_joinstyle='miter', # 선의 연결 모양, miter, round, bevel
    
)
```

```python
# title, xlabel, ylabel, legend 메서드는 그래프 그리기 전에 호출해야 함
# title, label : 기본적으로 한글 폰트는 지원하지 않음
# 한글 폰트 사용 시 fontproperties 인자에 한글 폰트 지정

plt.plot(x, y)

plt.title('Hello Matplotlib, 한글 설정') # 그래프의 제목
plt.xlabel('X') # x축 레이블
plt.ylabel('Y') # y축 레이블
plt.legend(loc='best') # 범례 표시, 범례에 표시될 이름은 label 인자에 지정한 이름
plt.show()
```

```python
plt.figure(figsize=(6, 2))

x = np.arange(1, 101)
y = np.linspace(2, 200, 100)

plt.grid(True) # 그래프에 격자 표시
plt.grid(color='red',
         linestyle='--',
         linewidth=0.5,
         axis='y', # 격자 표시 축 설정, axis='both', 'x', 'y'
         

         alpha=0.5,
        #  zorder=1000 # ? 
         ) # 격자 색상, 선 스타일, 선 두께 설정

# lim : 그래프의 범위 설정
plt.xlim(-10, 200) # x축 범위 설정
plt.ylim(-50, 400) # y축 범위 설정

plt.xticks(ticks=np.arange(0, 201, 50), fontsize=10, rotation=75) # x축 눈금 설정
plt.yticks(ticks=np.arange(0, 401, 100), fontsize=10, rotation=-30) # y축 눈금 설정

plt.plot(x, y)
plt.show()
```

```python
# lengend(범례)
plt.figure(figsize=(6, 4))

x = np.arange(1, 101)
y = np.linspace(50, 150, 100)
y2 = np.linspace(150, 50, 100)

plt.plot(x, y, label='시간')
plt.plot(x, y2, label='공간')
plt.legend(loc='best')
# loc 범례 위치 (기본 값 : best)
# best : 자동으로 범례 위치 선택
# - (upper | lower) (right | left)
# - 범례 위치 지정 시 범례 위치 지정


# 범례 표시, 범례에 표시될 이름은 label 인자에 지정한 이름
# loc : 범례 위치, best, upper right, upper left, lower left, lower right, right, center left, center right, lower center, upper center, center
plt.title('시간과 공간의 관계')
plt.xlabel('시간')
plt.ylabel('공간')
plt.show()

```

```python
##### 그래프 여러 개 그리기리기
```

```python
plt.figure(figsize=(6, 4))

x = np.arange(1, 11)
y = x ** 2
y2 = x ** 3

plt.plot(x, y, label='선', color='red', marker='o')
plt.bar(x, y2, label='바', color='green')

plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
# subplots 이용
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(6, 4))
ax[0].plot(x, y, label='선', color='red', marker='o')
ax[1].bar(x, y2, label='바', color='green')
ax[0].legend(loc='best')
ax[1].legend(loc='best')
ax[0].grid(True)
ax[1].grid(True)
plt.show()
```

```python
# subplots 이용
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
print(ax)

x1 = np.arange(1, 11)
y1 = x1 ** 4

x2 = np.arange(11, 19)
y2 = x2 ** 2

ax[0].plot(x1, y1, label='선', color='red', marker='o')
ax[0].legend(loc='best')
ax[0].set_title('선 그래프')
ax[0].set_xlabel('X1')
ax[0].set_ylabel('Y1')
ax[0].grid(True)

ax[1].bar(x2, y2, label='바', color='yellowgreen')
ax[1].legend(loc='best')
ax[1].set_title('바 그래프')
ax[1].set_xlabel('X2')
ax[1].set_ylabel('Y2')
ax[1].grid(True)

# 그래프 간격 조정
plt.tight_layout() # 그래프 간격 조정
plt.show()
```
